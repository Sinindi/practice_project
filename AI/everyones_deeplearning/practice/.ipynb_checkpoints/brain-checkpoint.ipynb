```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import optimizers, initializers, regularizers, metrics

np.random.seed(3)
tf.random.set_seed(3)


"""
학습셋에 이미지를 부풀리기 위해 여러 설정을 넣는다.
머신러닝이나 딥러닝에서 학습을 시키기위해선 많은 양의 데이터를 가지고
학습을 시킬수록 더 양질의 결과를 획득할 수 있다.
하지만 실제로 개인이 습득할 수 있는 양의 데이터는 한정적 혹은 적은양이기 때문에
이미지를 확대 축소, 반전, 회전 등으로 데이터를 부풀려 학습셋에 추가하는 방식으로
이미지 데이터를 늘리는 방식으로 부족한 양의 데이터로 학습을 진행 할 수 있다.
하지만 목적에 따라 무의미한 반전 회전 확대 축소는 과적합만을 발생 시킬 수 있기 때문에
목적에 맞게 적합한 방식의 이미지 변형을 수행하여야만 양질의 결과를 얻을 수 있다.
"""
train_datagen = ImageDataGenerator(rescale=1./255, #이미지의 색은 0~255의 RGB값을 가지기때문에 255로나누어 0~1사이의 값으로
                                   horizontal_flip=True,# 수평 대칭 이미지를 50% 확률로 만들어 추가
                                   width_shift_range=0.1, #전체 크기의 10%범위에서 좌우로 이동
                                   height_shift_range=0.1, #전체 크기의 10%범위에서 상하로 이동
                                   fill_mode='nearest') 
"""
fill_mode는 이미지를 축소 또는 회전하거나 이동할 때 생기는 빈 공간을 어떻게 채울지 결정하는 환경변수이다.
nearset 옵션을 선택하면 가장 비슷한 색으로 채워집니다.
이외에도 
rotation_range=5, 정해진 각도만큼 이미지를 회전시킵니다.
shear_range=0.7, 좌표하나늘 고정시키고 다른 몇개의 좌표를 이동시킨느 변환
zoom_range=[0.9,2.2], 정해진 범위 안에서 축소 또는 확대
vertical_filp=True 수직으로 반전
의 옵션이 있다.
"""

"""
앞서 수행한 학습과 다르게 이번 학습은 가지고 있는 이미지로 학습셋 테스트셋을 불러오기 때문에
불러오는 과정에서 datagen을 수행한다.
"""
train_generator = train_datagen.flow_from_directory('./train', #이미지가 위치한 폴더 위치
                                                    target_size=(150, 150), #이미지 크기
                                                    batch_size=5,
                                                    class_mode='binary') # 치매/정상 2진 분류이므로 바이너리 모드로 실행
"""
테스트 셋은 과적합을 방지 하기 위해서 색변형 이외의 변형은 하지 않는다.
"""
test_datagen = ImageDataGenerator(rescale = 1./255)


"""
앞서 학습셋을 불러오는것과 동일한 방식으로 테스트셋을 불러온다.
"""
test_generator = test_datagen.flow_from_directory('./test',
                                                  target_size = (150, 150),
                                                  batch_size = 5,
                                                  class_mode = 'binary')
model = Sequential()
model.add(Conv2D(32, (3,3), input_shape=(150, 150, 3 )))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(2))
model.add(Activation('sigmoid'))

model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizers.Adam(learning_rate=0.0002), metrics=['accuracy'])

history = model.fit(train_generator, #train_generator를 학습 모델로 사용
                    steps_per_epoch=32, #이미지 생성기에서 몇개의 샘플을 뽑을지 결정
                    epochs=20, 
                    validation_data=test_generator, #앞서 만들어진 test_generator를 테스트셋으로 사용
                    validation_steps=4)

acc= history.history['accuracy']
val_acc = history.history['val_accuracy']
y_vloss = history.history['val_loss']
y_loss = history.history['loss']

x_len = np.arange(len(y_loss))
plt.plot(x_len, acc, marker='.', c='red', label='Trainset_acc')
plt.plot(x_len, val_acc, marker='.', c='lightcoral', label='Testset_acc')
plt.plot(x_len, y_vloss, marker='.', c="cornflowerblue", label='Testset_loss')
plt.plot(x_len, y_loss, marker='.', c="blue", label='Trainset_loss')

plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss/acc')
plt.show()
```
