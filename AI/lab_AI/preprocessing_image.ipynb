```python
import os,sys
from PIL import Image

current_path = os.path.abspath('')
image_path = os.path.join(current_path, "test_img")
modified_path = os.path.join(current_path, "test_img2")


"""
이미지를 재단하는 함수
"""
def crop_img(image_path,modified_path):
    
    
    files = os.listdir(image_path)
    select = re.compile("([a-zA-Z0-9_\s\.]+(?=\_pic))")
    
    for file in files:
        if select.match(str(file)) == None:
            name = str(file)
            os.chdir(image_path)
            img = Image.open(file)

            img = img.crop([136, 112, 917, 823])

            os.chdir(modified_path)
            img.save(name, "PNG")
            os.chdir('..')
        else:
            name = str(file)
            os.chdir(image_path)
            img = Image.open(file)

            img = img.crop([216, 289.5, 836, 645])

            os.chdir(modified_path)
            img.save(name, "PNG")
            os.chdir('..')
        
"""
주피터는 vscode와 다르게 함수가 실행된 후의 진행상황이 보관되는거같다.
그래서 디렉토리 변경 로그가 남아 한번 실행하면 더이상 사용못하는 상황이되서
초기화 과정이 필수적으로 들어가야만 하는거 같다.
"""
crop_img(image_path,modified_path)
```

```python
import numpy as np
import re

"""
정규식이 들어간이유
현재 저장된 이미지는 회색이미지와 색이 들어간 이미지가 함께 들어가있다
회색이미지는 rgb값이 필요가 없기때문에 이 방식으로 전처리하지않고
다른 방법으로 전처리 할 예정이다.
"""
def img_preprocessing(image_path):
    #배열 선언 및 초기화
    save = []
    files = os.listdir(image_path)
    #정규식 _pic이 포함이 안되있으면 none을 배출한다.
    select = re.compile("([a-zA-Z0-9_\s\.]+(?=\_pic))")
    
    for file in files:
        if select.match(str(file)) == None:
            os.chdir(image_path)
            #패스에서 이미지 불러오기
            img = Image.open(file)
            #이미지를 배열로 변환
            img_tensor = image.img_to_array(img)
            print(img_tensor.shape)
            #배열로 변환된 이미지는 (x픽셀, y픽셀, rgb)의 형태로 재정의된다.
            #3차원 배열을 1차원 배열로 변환 시키기 위해 reshape함수를 이용하여 변환
            #변환후 값을 0~1사이의 값으로 변환하기 위해 255로 나누어준다.
            img_train = img_tensor.reshape(1665873).astype('float64')/ 255
            #생성된 배열을 미리 정의해둔 배열에 추가
            save.append(img_train)
            #경로 초기화
            os.chdir('..')
        else:
            continue
            
    #생성된 배열을 csv파일 형식으로 현재 디렉토리에 저장한다.
    np.savetxt('saved.csv', save, delimiter=',')
        
img_preprocessing(modified_path)
```

```python
def img_preprocessing_binary(image_path):
    #배열 선언 및 초기화
    save = []
    binary_img_path = os.path.join(current_path,"binary")
    files = os.listdir(image_path)
    #정규식 _pic이 포함이 안되있으면 none을 배출한다.
    select = re.compile("([a-zA-Z0-9_\s\.]+(?=\_pic))")
    
    for file in files:
        if select.match(str(file)) != None:
            os.chdir(image_path)
            img_color  = cv2.imread(file, cv2.IMREAD_COLOR)
            img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
            ret, img_binary = cv2.threshold(img_gray, 212, 255, cv2.THRESH_BINARY)
            img_train = img_binary.reshape(220100).astype('float64')/ 255
            #생성된 배열을 미리 정의해둔 배열에 추가
            save.append(img_train)
            os.chdir(binary_img_path)
            cv2.imwrite(str(file),img_binary)
            #경로 초기화
            os.chdir('..')
        else:
            continue
            
    #생성된 배열을 csv파일 형식으로 현재 디렉토리에 저장한다.
    np.savetxt('saved_binary.csv', save, delimiter=',')
        
img_preprocessing_binary(modified_path)
```
