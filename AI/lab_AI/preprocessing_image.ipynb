```python
def img_preprocessing_binary(image_path):
    #   
    save = []
    binary_img_path = os.path.join(current_path,"binary")
    files = os.listdir(image_path)
    # _pic   none .
    select = re.compile("([a-zA-Z0-9_\s\.]+(?=\_pic))")
    
    for file in files:
        if select.match(str(file)) != None:
            os.chdir(image_path)
            img_color  = cv2.imread(file, cv2.IMREAD_COLOR)
            img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
            ret, img_binary = cv2.threshold(img_gray, 212, 255, cv2.THRESH_BINARY)
            img_train = img_binary.reshape(220100).astype('float64')/ 255
            #     
            save.append(img_train)
            os.chdir(binary_img_path)
            cv2.imwrite(str(file),img_binary)
            # 
            os.chdir('..')
        else:
            continue
            
    #  csv    .
    np.savetxt('saved_binary.csv', save, delimiter=',')
        
img_preprocessing_binary(modified_path)
```

```python
def img_preprocessing_edge(image_path):
    #   
    save = []
    edge_img_path = os.path.join(current_path,"edge")
    files = os.listdir(image_path)
    # _pic   none .
    select = re.compile("([a-zA-Z0-9_\s\.]+(?=\_pic))")
    
    for file in files:
        if select.match(str(file)) != None:
            os.chdir(image_path)
            img_color  = cv2.imread(file, cv2.IMREAD_COLOR)
            img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
            canny = cv2.Canny(img_gray, 100, 255)
            img_train = canny.reshape(220100).astype('float64')/ 255
            #     
            save.append(img_train)
            os.chdir(edge_img_path)
            cv2.imwrite(str(file),canny)
            # 
            os.chdir('..')
        else:
            continue
            
    #  csv    .
    np.savetxt('saved_edge.csv', save, delimiter=',')
        
img_preprocessing_edge(modified_path)
```
